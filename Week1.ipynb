step 1

import numpy, pandas, cv2, skimage, dan matplotlib

```
import numpy as np
import pandas as pd
import cv2 as cv
from google.colab.patches import cv2_imshow #image display
from skimage import io
from skimage import transform
from PIL import Image
import matplotlib.pylab as plt
```

step 2 

membaca image dari url, menampilkan opencv

```
#list menyimpan url dari beberapa image
urls = ["https://iiif.lib.ncsu.edu/iiif/0052574/full/800,/0/default.jpg",
"https://iiif.lib.ncsu.edu/iiif/0016007/full/800,/0/default.jpg",
"https://placekitten.com/800/777"]
#baca dan tamplikan image
#loop pada tiap url image, beberapa imae dapat disipan di list
for url in urls:
  image = io.imread(url) #membaca gambar
  image = cv.resize(image, (0,0), fx=0.5, fy=0.5) #resize gambar menjadi half size
  image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB) #convert warna ke rgb
  final_frame = cv.hconcat((image, image_2)) #concatenate img
  cv2_imshow(final_frame) # show gambar
  print('\n')
```

step 3

mwlihat ukuran gambar

```
tinggi = image_2.shape[0]
lebar = image_2.shape[1]
print("resolusi image: tinggi x lebar = ",tinggi," x ",lebar)
cv2_imshow(image_2)
```

step 4

mengakses pixel dengan memberikan garis
horizontal berwarna putih di tengah image

```
image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_3 = cv.cvtColor(image, cv.COLOR_BGR2RGB)

#membuat garis horizontal
for y in range (lebar):
  image_3[int((tinggi)/2),y] = [0,0,0]

final_frame = cv.hconcat((image_2, image_3))
cv2_imshow(final_frame)
```

garis menyilang

```
image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_3 = cv.cvtColor(image, cv.COLOR_BGR2RGB)

# Membuat garis diagonal ditengah image
image_3 = cv.line(image_3,(0,0), (lebar,tinggi), [0,0,0,0])
image_3 = cv.line(image_3,(lebar, 0), (0,tinggi), [0,0,0,0])

final_frame = cv.hconcat((image_2, image_3))
cv2_imshow(final_frame)
```
